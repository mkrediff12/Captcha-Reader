```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('/kaggle/input/train/ANMH.png')
plt.imshow(img)
```

```python
# Convert the image into RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   #default setting of the color mode in OpenCV comes in the order of BGR, which is different from that of Matplotlib.
                                                 #Therefore to see the image in RGB mode, we need to convert it from BGR to RGB as follows
plt.imshow(img_rgb)
```

```python
#converting image to gray scale and then thresholding is done
img_gray=cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray,cmap='gray')
```

```python
_, thresh_img = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh_img,cmap='gray')
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
morph_img = cv2.morphologyEx(thresh_img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(morph_img,cmap='gray')
```

```python
# Find the contour of the figure 
    contours, hierarchy = cv2.findContours(
                                   image = morph_img, 
                                   mode = cv2.RETR_TREE, 
                                   method = cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Sort the contours 
# contours = sorted(contours, key = cv2.contourArea, reverse = True)
# Draw the contour 
img_copy = img_rgb.copy()
final = cv2.drawContours(img_copy, contours, contourIdx = -1, 
                         color = (255, 0, 0), thickness = 2)
# img_rgb = cv2.cvtColor(img_copy, cv2.COLOR_GRAY2RGB) 
plt.imshow(img_copy)
```

```python
contours_new=[]
threshold_area = 1400     #threshold area    
for cnt in contours:        
    area = cv2.contourArea(cnt)         
    if area > threshold_area:                   
         contours_new.append(cnt)
```

```python
# Sort the contours 
# contours = sorted(contours, key = cv2.contourArea, reverse = True)
# Draw the contour 
img_copy = img_rgb.copy()
final = cv2.drawContours(img_copy, contours_new, contourIdx = -1, 
                         color = (255, 0, 0), thickness = 2)
# img_rgb = cv2.cvtColor(img_copy, cv2.COLOR_GRAY2RGB) 
plt.imshow(img_copy)
```
